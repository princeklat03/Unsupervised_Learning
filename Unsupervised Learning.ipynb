```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from sklearn.cluster import KMeans
```

```python
plt.imshow(imag)
```

```python
imag_array_re = imag_array.reshape(imag_array.shape[0]*imag_array.shape[1], imag_array.shape[2])
```
